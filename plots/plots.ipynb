```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
sns.set_palette("deep")
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
df.head()
```

```python
fig = sns.catplot(data = df, x = 'stroke', y = 'bmi', hue = 'heart_disease', kind = 'strip')
fig.savefig('bmi_vs_stroke_other.jpg')
```

```python
rate = []
for i in range(int(df.age.min()), int(df.age.max())):
    rate.append(df[df['age'] < i]['stroke'].sum() / len(df[df['age'] < i]['stroke']))

fig = sns.lineplot(data = rate)
fig.set(xlabel='age', ylabel='rate of stroke')
plt.savefig('age_vs_strokerate.jpg')
```

```python
sns.catplot(data = df, x = 'gender', kind = 'count', hue = 'stroke')
```

```python
sns.set(rc={'figure.figsize':(50,50)})
sns.pairplot(df)
plt.savefig("pairplot.jpg")
```
